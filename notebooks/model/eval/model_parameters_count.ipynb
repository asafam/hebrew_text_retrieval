```python
from transformers import AutoModel

def get_model_size(model_name):
    model = AutoModel.from_pretrained(model_name)
    total_params = sum(p.numel() for p in model.parameters())
    return total_params
model = AutoModel.from_pretrained("bert-base-uncased")

model_name = "bert-base-uncased"
total_params = get_model_size(model_name)
print(f"Total parameters: {total_params:,}")
```

```python
from tqdm import tqdm

model_names = [
    "onlplab/alephbert-base",
    "dicta-il/dictabert",
    "intfloat/multilingual-e5-base",
    "intfloat/multilingual-e5-large",
    "/home/nlp/achimoa/workspace/ModernBERT/hf/HebrewModernBERT/ModernBERT-Hebrew-base_20250622_1325/ep7-ba896339-rank0",
]

for model_name in tqdm(model_names):
    total_params = get_model_size(model_name)
    print(f"Model: {model_name}, Total parameters: {total_params:,}")
```
